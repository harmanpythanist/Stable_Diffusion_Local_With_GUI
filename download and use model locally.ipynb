## Saving the model

```python
from diffusers import StableDiffusionPipeline
import torch

model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float32
)

# Save to D drive
pipe.save_pretrained("D:/models/sd_v1_5_local")
```

## Loading model

```python
import torch
from diffusers import StableDiffusionPipeline

pipe = StableDiffusionPipeline.from_pretrained(
    "D:/models/sd_v1_5_local",   # local path
    torch_dtype=torch.float32,
    local_files_only=True
)
```

```python
pipe = pipe.to("cuda")
pipe.safety_checker = None

prompt = "two years old girl in her mother's lap, sitting on a bench, father sitting beside using his mobile, in a clinic"

image = pipe(prompt, num_inference_steps=25).images[0]
image.save("gpu_result_3.png")
```
